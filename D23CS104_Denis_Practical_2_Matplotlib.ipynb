**1) Creating a Chart**

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.arange(0, 20)
y = x ^ 3
# Simple Plot
plt.plot(x, y)
```

**2) Labling the Axes**

```python
plt.title("Cubic relationship between x and y")   #Labeling the Axes and Title
plt.xlabel("Time")
plt.ylabel("Distance")
plt.plot(x, y)
```

**3) Formatting Line Type and Color**

```python
import numpy as np
import matplotlib.pyplot as plt
x = np.arange(0, 20)
y = x ^ 3
plt.title("Cubic relationship between x and y")   #Labeling the Axes and Title
plt.xlabel("Time")
plt.ylabel("Distance")
plt.plot(x, y, "r") #Formatting the Line colors
plt.plot(x, y, ">") #Formatting the line type
```

**4) Saving Graph**

```python
import numpy as np
import matplotlib.pyplot as plt
x = np.arange(0, 20)
y = x ^ 3
plt.title("Cubic relationship between x and y")   #Labeling the Axes and Title
plt.xlabel("Time")
plt.ylabel("Distance")
plt.plot(x, y, "r") #Formatting the Line colors
plt.plot(x, y, ">") #Formatting the line type
plt.savefig("timevsdist.png", format="png", dpi=300) #we can also save in pdf formats
```

**5) Adding Annotations**

```python
import numpy as np
import matplotlib.pyplot as plt
x = np.arange(0, 10)
y = x ^ 2
z = x ^ 3
t = x ^ 4

# Labeling the axes and title
plt.title("Graph Drawing")
plt.xlabel("Time")
plt.ylabel("Distance")
plt.plot(x, y)
plt.annotate(xy=[2, 1], text="Second Entry")
plt.annotate(xy=[4, 6], text="Third Entry")
```

**6) Adding Legends**

```python
import numpy as np
import matplotlib.pyplot as plt
x = np.arange(0, 10)
y = x ^ 2
z = x ^ 3
t = x ^ 4

# Labeling the axes and title
plt.title("Graph Drawing")
plt.xlabel("Time")
plt.ylabel("Distance")
plt.plot(x, y)
plt.annotate(xy=[2, 1], text="Second Entry")
plt.annotate(xy=[4, 6], text="Third Entry")

# Adding Legends
plt.plot(x, z)
plt.plot(x, t)
plt.legend(['Race1', 'Race2', 'Race3'], loc=4)
```

**7) Superimposing 2 Graphs**

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
def f(t):
    return np.exp(-t) * np.cos(2*np.pi*t)
t1 = np.arange(0.0, 5.0, 0.1)
t2 = np.arange(0.0, 5.0, 0.02)
plt.figure(1)
plt.plot(t1, f(t1), '*', t2, f(t2), 'r')
```

**8) Plotting Pie Graph**

```python
import matplotlib.pyplot as plt
labels = 'Frogs', 'Hogs', 'Dogs', 'Logs'
sizes = [15, 30, 45, 10]
explode = (0, 0, 0.2, 0)  # only "explode" the 2nd slice (i.e. 'Hogs')
fig1, ax1 = plt.subplots()
ax1.pie(sizes, explode=explode, labels=labels, autopct='%1.1f%%',
        shadow=True, startangle=90)
ax1.axis('equal')           # Equal aspect ratio ensures that pie is drawn as a circle
plt.legend(['Frogs', 'Hogs', 'Dogs', 'Logs'], loc=0)
plt.show()
```

**10) Subplots**

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
x = np.linspace(0, 5, 11)
y = x ** 2
plt.subplot(1, 2, 1)
plt.plot(x, x, 'r--')
plt.subplot(1, 2, 2)
plt.plot(y, x, 'g*-')
plt.xlabel("X Axis")
plt.ylabel("Y Axis")
plt.title("Demo Title")
```

**11) Another Way Of Using Subplots**

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
x = np.linspace(0, 5, 11)
fig, ax = plt.subplots()

# generate 2 objects, fig and ax
ax.plot(x, x**2, "b.-")
ax.plot(x, x**3, "g--")
ax.plot(x, x**4, "r")
ax.plot(x, x**5, "y*")
ax.plot(x, x**6, "black")

plt.xlabel("X Axis")
plt.ylabel("Y Axis")
plt.title("Demo Title")
```

**12) Histogram With Text**

```python
mu, sigma = 100, 15
x = mu + sigma * np.random.randn(10000)
# the histogram of the data
n, bins, patches = plt.hist(x, 50, density=1, facecolor='b', alpha=0.75)
plt.xlabel('Smarts')
plt.ylabel('Probability')
plt.title('Histogram of IQ')
plt.text(60, .025, r'$\mu=100,\ \sigma=15$')
plt.axis([40, 160, 0, 0.03])
plt.grid(True)
plt.show()
```

**13) Plot Different variety of graphs as subplots**

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

n = np.array([0,1,2,3,4,5])
xx = np.linspace(-0.75, 1., 100)
fig, axes = plt.subplots(1, 4, figsize=(12,3))
axes[0].scatter(xx, xx + 0.25*np.random.randn(len(xx)))
axes[0].set_title("scatter")

axes[1].step(n, n**2, lw=2)
axes[1].set_title("step")

axes[2].bar(n, n**2, align="center", width=0.5, alpha=0.5)
axes[2].set_title("bar")

axes[3].fill_between(xx, xx**2, xx**3, color="green", alpha=0.5);
axes[3].set_title("fill_between");
```
