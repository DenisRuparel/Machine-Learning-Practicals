```python
import pandas as pd
import numpy as np
from sklearn import linear_model
import matplotlib.pyplot as plt

df = pd.read_csv('/content/drive/MyDrive/Salary_dataset.csv')
df
```

```python
plt.xlabel('YearsExperience')
plt.ylabel('Salary')
plt.scatter(df.YearsExperience,df.Salary,color='red',marker='+')
```

```python
new_df = df.drop('Salary', axis='columns')
new_df

x= new_df.drop('Unnamed: 0',axis=1)
x

Salary = df.Salary
Salary
# # # Create linear regression object
reg = linear_model.LinearRegression()
reg.fit(x, Salary)
reg.predict([[10]])
```

```python
plt.xlabel('YearsExperience')
plt.ylabel('Salary')
plt.scatter(df.YearsExperience,df.Salary,color='red',marker='+')
plt.plot(df.YearsExperience,reg.predict(df[['YearsExperience']]),color='blue')
```
