```python
import pandas as pd
import numpy as np
from sklearn import linear_model
import matplotlib.pyplot as plt

df = pd.read_csv('/content/drive/MyDrive/MyDataset.csv')
df
```

```python
plt.xlabel('CarsName')
plt.ylabel('Price')
plt.scatter(df.Car_Name,df.Selling_Price,color='blue',marker='+')
```
